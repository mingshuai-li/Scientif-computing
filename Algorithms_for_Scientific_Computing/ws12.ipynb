# Arithmetization of Space-Filling Curves

If the number $t \in [0,1[$ is given with the basis four, i.e.

\begin{equation}
   t = 0_4.q_1 q_2 q_3 q_4 \ldots,
\end{equation}
then the mapping $h(t)$ of the Hilbert curve can be written as
\begin{equation}
   h(0_4.q_1 q_2 q_3 q_4 \ldots) =
   H_{q_1} \circ H_{q_2} \circ H_{q_3} \circ H_{q_4} \circ \cdots
   \left( 0 \atop 0 \right)
\end{equation}
with the operators
\begin{equation}
 \begin{array}{rclrcl}
   H_0 &:=& \left( \begin{array}{cc} 
                      0 & \frac{1}{2} \\ 
                      \frac{1}{2} & 0 \end{array} \right)  \displaystyle
            \left( x \atop y \right) 
   &
   H_1 &:=& \left( \begin{array}{cc} 
                      \frac{1}{2} & 0 \\ 
                      0 & \frac{1}{2} \end{array} \right) \displaystyle
            \left( x \atop y \right) + \left( 0 \atop \frac{1}{2} \right) 
   &
   H_2 &:=& \left( \begin{array}{cc} 
                      \frac{1}{2} & 0 \\ 
                      0 & \frac{1}{2} \end{array} \right)  \displaystyle
            \left( x \atop y \right) + \left( \frac{1}{2} \atop \frac{1}{2} \right)
   &
   H_3 &:=& \left( \begin{array}{cc} 
                      0 & -\frac{1}{2} \\ 
                      -\frac{1}{2} & 0 \end{array} \right)  \displaystyle
            \left( x \atop y \right)  + \left( 1 \atop \frac{1}{2} \right). 
 \end{array}
\end{equation}

# Exercise 2: Arithmetization of the Peano Curve

Derive an arithmetization of the Peano curve (see figure \ref{fig:peano}a), analog to the arithmetization of the Hilbert curve. So we are looking for a representation like

\begin{equation}
   p(0.q_1 q_2 q_3 q_4 \ldots) =
   P_{q_1} \circ P_{q_2} \circ P_{q_3} \circ P_{q_4} \circ \cdots
   \left( 0 \atop 0 \right)
\end{equation}

constructed on an appropriate basis and with appropriate operators
$P_0$, $P_1$, $\dots$

Develop an algorithm that computes the Peano function $p(t)$. 

Use the Peano function to plot the Peano curve (with Python).

```python
# include plotting functionality
import matplotlib.pyplot as plt
from math import floor

def plotLineUp(x, y, length):
    x.append(x[len(x) - 1])
    y.append(y[len(y) - 1] + length)

def plotLineStrip(x, y, title = None):
    _, ax = plt.subplots()
    plt.plot(x, y)
    if title is not None:
        plt.title(title)
    # unit square with some extra space around
    ax.set_xlim(-0.05, 1.05)
    ax.set_ylim(-0.05, 1.05)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
# arithmetization of the Hilbert curve
def H0(pos):
    x = pos[1] * 0.5
    y = pos[0] * 0.5
    return (x, y)

def H1(pos):
    x = pos[0] * 0.5
    y = pos[1] * 0.5 + 0.5
    return (x, y)

def H2(pos):
    x = pos[0] * 0.5 + 0.5
    y = pos[1] * 0.5 + 0.5
    return (x, y)

def H3(pos):
    x = -pos[1] * 0.5 + 1.0
    y = -pos[0] * 0.5 + 0.5
    return (x, y)
```

```python
# arithmetization of the Peano curve
def P0(pos):
    x = pos[0] / 3.0
    y = pos[1] / 3.0
    return (x, y)

def P1(pos):
    x = -pos[0] / 3.0 + 1.0 / 3.0
    y = pos[1] / 3.0 + 1.0 / 3.0
    return (x, y)

def P2(pos):
    x = pos[0] / 3.0
    y = pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)

def P3(pos):
    x = pos[0] / 3.0 + 1.0 / 3.0
    y = -pos[1] / 3.0 + 1.0
    return (x, y)

def P4(pos):
    x = -pos[0] / 3.0 + 2.0 / 3.0
    y = -pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)

def P5(pos):
    x = pos[0] / 3.0 + 1.0 / 3.0
    y = -pos[1] / 3.0 + 1.0 / 3.0
    return (x, y)

def P6(pos):
    x = pos[0] / 3.0 + 2.0 / 3.0
    y = pos[1] / 3.0
    return (x, y)

def P7(pos):
    x = -pos[0] / 3.0 + 1.0
    y = pos[1] / 3.0 + 1.0 / 3.0
    return (x, y)

def P8(pos):
    x = pos[0] / 3.0 + 2.0 / 3.0
    y = pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)
```

```python
# arithmetization of the meander-type Peano curve
def M0(pos):
    x = pos[1] / 3.0
    y = pos[0] / 3.0
    return (x, y)

def M1(pos):
    x = pos[1] / 3.0
    y = pos[0] / 3.0 + 1.0 / 3.0
    return (x, y)

def M2(pos):
    x = pos[0] / 3.0
    y = pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)

def M3(pos):
    x = pos[0] / 3.0 + 1.0 / 3.0
    y = pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)

def M4(pos):
    x = pos[0] / 3.0 + 2.0 / 3.0
    y = pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)

def M5(pos):
    x = -pos[0] / 3.0 + 1.0
    y = -pos[1] / 3.0 + 2.0 / 3.0
    return (x, y)

def M6(pos):
    x = -pos[1] / 3.0 + 2.0 / 3.0
    y = -pos[0] / 3.0 + 2.0 / 3.0
    return (x, y)

def M7(pos):
    x = -pos[1] / 3.0 + 2.0 / 3.0
    y = -pos[0] / 3.0 + 1.0 / 3.0
    return (x, y)

def M8(pos):
    x = pos[0] / 3.0 + 2.0 / 3.0
    y = pos[1] / 3.0
    return (x, y)
```

```python
## set the number of the program you want to test
# 1 = Hilbert; 2 = Peano; 3 = Peano-Meander
prog = 3

## depth of the curve
depth = 2


t_start = 0.0
t_end = 1.0
x = []
y = []

# plot the hilbert curve
if prog == 1:
    while t_start < t_end:
        hilbert = []
        pos = (0.5, 0.5) #start at (0.5,0.5) to get the point in the middle of the cell
        temp = t_start
        for i in range(depth):
            q = floor(4 * temp)
            temp = 4 * temp - q
            hilbert.append(q)
        for i in range(len(hilbert)):
            if hilbert[len(hilbert)-i-1] == 0:
                pos = H0(pos)
            elif hilbert[len(hilbert)-i-1] == 1:
                pos = H1(pos)
            elif hilbert[len(hilbert)-i-1] == 2:
                pos = H2(pos)
            elif hilbert[len(hilbert)-i-1] == 3:
                pos = H3(pos)
        x.append(pos[0])
        y.append(pos[1])
        t_start = t_start + 0.25**depth

    plotLineStrip(x, y, None)

# plot the peano curve
elif prog == 2:
    eps = (1.0 / 9.0)**depth

    for k in range(0, 9**depth):
        peano = []
        pos = (0.5, 0.5)  #start at (0.5,0.5) to get the point in the middle of the cell
        temp = k*eps
        for i in range(depth):
            q = floor(9*temp + eps/2) #add eps/2 due to numeric issues
            temp = 9 * temp - q
            peano.append(q)
        for i in range(len(peano)):
            if peano[len(peano)-i-1] == 0:
                pos = P0(pos)
            elif peano[len(peano)-i-1] == 1:
                pos = P1(pos)
            elif peano[len(peano)-i-1] == 2:
                pos = P2(pos)
            elif peano[len(peano)-i-1] == 3:
                pos = P3(pos)
            elif peano[len(peano)-i-1] == 4:
                pos = P4(pos)
            elif peano[len(peano)-i-1] == 5:
                pos = P5(pos)
            elif peano[len(peano)-i-1] == 6:
                pos = P6(pos)
            elif peano[len(peano)-i-1] == 7:
                pos = P7(pos)
            elif peano[len(peano)-i-1] == 8:
                pos = P8(pos)
        x.append(pos[0])
        y.append(pos[1])
    plotLineStrip(x, y, None)

# plot approximatin polygon for a Peano-Meander curve
elif prog == 3:
    eps = (1.0 / 9.0)**depth

    for k in range(0, 9**depth):
        peano = []
        pos = (0.5, 0.5) #start at (0.5,0.5) to get the point in the middle of the cell
        temp = k*eps
        last_temp = 1.0 - temp
        for i in range(depth):
            q = floor(9*temp + eps/2) #add eps/2 due to numeric issues
            temp = 9 * temp - q
            peano.append(q)
        for i in range(len(peano)):
            if peano[len(peano)-i-1] == 0:
                pos = M0(pos)
            elif peano[len(peano)-i-1] == 1:
                pos = M1(pos)
            elif peano[len(peano)-i-1] == 2:
                pos = M2(pos)
            elif peano[len(peano)-i-1] == 3:
                pos = M3(pos)
            elif peano[len(peano)-i-1] == 4:
                pos = M4(pos)
            elif peano[len(peano)-i-1] == 5:
                pos = M5(pos)
            elif peano[len(peano)-i-1] == 6:
                pos = M6(pos)
            elif peano[len(peano)-i-1] == 7:
                pos = M7(pos)
            elif peano[len(peano)-i-1] == 8:
                pos = M8(pos)
        x.append(pos[0])
        y.append(pos[1])
    plotLineStrip(x, y, None)
else:
    exit(1)
    print("Error: Invalid exercise number", prog)
```
