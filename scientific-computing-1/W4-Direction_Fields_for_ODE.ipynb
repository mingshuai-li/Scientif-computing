# Direction Fields of ODE's

Sample script to plot the direction fields for Worsheet 4, Assignment 1

### Imports and Helper Functions

```python
import matplotlib.pyplot as plt
%matplotlib inline  
import numpy as np

def plotDirectionField(X, Y, U, V, pname):
    '''
    Plotting of the direction field. The "quiver" function displays the vectors on the plot.
    
    Arguments:
        X,Y: the midpoint of the arrow
        U,V: the direction of the arrow
        pname: Title of the plot
    '''
    plt.figure(figsize=(9,6))
    M = np.sqrt(pow(U, 2) + pow(V, 2))
    U = np.divide(U,M)
    V = np.divide(V,M)
    Q = plt.quiver( X, Y, U, V, units='xy', pivot='mid', width=0.06, scale=1/.5)
    plt.title(pname)
    plt.show()

def vectorField(y, p):
    '''
    Evaluates y' for the ODE 
        y'(t)=λ∗y(t)²+μ∗y(t)−ν
    at given points y.
        
    Arguments:
        y : vector of y-values to evaluate
        p : vecor of parameters [λ, μ, ν]
    '''
    lam, mu, nu = p
    return lam*Y*Y + mu*Y - nu
```

## Ploting the ODE
## $$y'(t) = \lambda y^2+\mu y-\nu$$

### Saddle Point:  $\mu$ = 0, $\nu$ = 0

Critical point at 0


```python
X,Y = np.meshgrid( np.arange(0,20.1,1.),np.arange(-1.,1.1,.1) )
lam = 1.
mu = 0.
nu = 0.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])


plotDirectionField(X, Y, U, V, "Saddle Point")
```

### Two critical points: $\lambda$ = 5, $\mu$ = 4, $\nu$ = 1

```python
X,Y = np.meshgrid( np.arange(0,20.1,1.),np.arange(-2.,1.1,.2) )
lam = 5.
mu = 4.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])

plotDirectionField(X, Y, U, V, "Two Critical Points")
```

### $\lambda$ = 0, $\mu$ = 1

We obtain $y(t) = \frac{\nu}{\mu}$. With positive $\mu$: Unstable point

```python

X,Y = np.meshgrid( np.arange(0,20.1,1.),np.arange(0,2.1,.1) )
lam = 0.
mu = 1.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])

plotDirectionField(X, Y, U, V, r"$\lambda = 0, \mu = 1, \nu=1$")
```

### $\lambda$ = 0, $\mu$ = -1

We obtain $y(t) = \frac{\nu}{\mu}$. With negative $\mu$: Stable point

```python

X,Y = np.meshgrid( np.arange(0,20.1,1.),np.arange(-2.0,-0.01,.1) )
lam = 0.
mu = -1.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])

plotDirectionField(X, Y, U, V, r"$\lambda = 0, \mu = -1, \nu=1$")
```

### $\lambda$= 0, $\mu$ = 0

if $\nu$ is different than zero, the equality $\frac{dy}{dt}$ = 0 can never be reached

```python
X,Y = np.meshgrid( np.arange(0,20.1,1.),np.arange(-1.0,1.1,.1) )
lam = 0.
mu = 0.
nu = 1.

U = np.copy(X)
U.fill(1)
V = vectorField(Y,[lam,mu,nu])

plotDirectionField(X, Y, U, V, r"$\lambda = 0, \mu = 0, \nu=1$")
```
