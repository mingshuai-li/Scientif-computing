```python
from pylab import *
import numpy as np
```

### Define a function for ploting the direction field

```python
def plotDirectionField(X, Y, U, V, pname):
    """ simple plotting of direction fields."""
    M = sqrt(pow(U, 2) + pow(V, 2))
    U = np.divide(U,M)
    V = np.divide(V,M)
    Q = quiver( X, Y, U, V, M, units='xy', pivot='mid', width=0.010, scale=1/0.08)
    title(pname)
    show()
```

```python
X,Y = meshgrid( arange(-1,1.08,.08),arange(-1,1.08,.08) )
```

## Ex. 1 e
$$\frac{d^2y}{dt^2}=-y$$

```python
U = Y
V = -X
figure(figsize=(7, 7), dpi=100)
plotDirectionField(X, Y, U, V, r"Direction Field of $\frac{d^2y}{dt^2}=-y$")
```

## Ex. 1 f
$$\frac{d^2y}{dt^2}=-\mu y$$

```python
mu = 0
U = Y
V = -mu*X
figure(figsize=(7, 7), dpi=100)
plotDirectionField(X, Y, U, V, r"Direction Field for $\frac{d^2y}{dt^2} =- \mu*y$, where $\mu =$ " + str(mu))
```
